## FLIPKART

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as req
```

```python
my_url = 'https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off'
```

```python
uclient = req(my_url)
```

```python
page_html = uclient.read()
```

```python
uclient.close()
```

```python
page_soup = soup(page_html,"html.parser")
```

```python
type(page_soup)
```

```python
page_soup
```

```python
container = page_soup.findAll("div",{"class":"_1UoZlX"})
```

```python
len(container)
```

```python
flipkart=[]
for i in range(24):
    
    flip_title = container[i].findAll("div",{"class":"_3wU53n"})
    flip_title1 = flip_title[0].text
    
    flip_final_price = container[i].findAll("div",{"class":"_1vC4OE _2rQ-NK"})
    flip_final_price1 = flip_final_price[0].text
    
    print(" title " + flip_title1)
    print(" final price " + flip_final_price1)
    
    flipkart.append([flip_title1,flip_final_price1])
    
```

```python
flipkart
```

```python
uclient.close()
```

```python
title1 = page_soup.findAll("p",{"class":"product-title"})
len(title1)
```

```python
Snapdeal=[]
for i in range(4):
    
    title1 = page_soup.findAll("p",{"class":"product-title"})
    title = title1[i].text
    
    Actual = page_soup.findAll("span",{"class":"lfloat product-price"})
    Actual = Actual[i].text[5:]
    
    
    print(" title " + title)
    print(" Actual Price " + Actual)
    Snapdeal.append([title,Actual])
```
